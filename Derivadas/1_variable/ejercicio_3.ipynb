### Ejercicios derivadas 1 variable con python

3) **Calcule la derivada de la función $f(x)=\frac{3}{x^2}$**

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
# Definicion de variables
x = sp.symbols('x')

# Definir la función
f = 3/x**2

# Calcular la derivada
derivada = sp.diff(f, x)

# Resultado
display(Math(r"f'(x) = " + sp.latex(derivada)))
```

```python
# Grafica de f(x) y f'(x)

# Conversion de calculo simbolico a numerico
import numpy as np

# Definir funciones
f_num = sp.lambdify(x, f, 'numpy')
f_x = sp.lambdify(x, derivada, 'numpy')

# Generar datos para graficar
X = np.linspace(-6, 6, 100)

# Evaluar funciones en puntos
F = f_num(X)
Fx = f_x(X)

# Plotear
plt.figure(figsize=(10, 6))
plt.plot(X, F, label=r"$f(x)=\frac{3}{x^2}$")   # Plotear f(x)
plt.plot(X, Fx, label=r"$f'(x)=-\frac{6}{x^{3}}$",linestyle="--")  # Plotear f'(x)

# Labels
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Escala de rango para y
plt.ylim(-30, 60)

# Título y leyendas
plt.title('Gráfica de $f(x)$ y $f\'(x)$', fontsize=16)
plt.legend(fontsize=12)

# Grilla
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```
