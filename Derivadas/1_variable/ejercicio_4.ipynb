### Ejercicios derivadas 1 variable con python

4) **Calcule la derivada de la función $f(x)=cos(3x+3)$**

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
# Definicion de variables
x = sp.symbols('x')

# Definir la función
f = sp.cos(3*x+3)

# Calcular la derivada
derivada = sp.diff(f, x)

# Resultado
display(Math(r"f'(x) = " + sp.latex(derivada)))
```

```python
# Grafica de f(x) y f'(x)

# Conversion de calculo simbolico a numerico
import numpy as np
f_num = sp.lambdify(x, f, 'numpy')
derivada_num = sp.lambdify(x, derivada, 'numpy')

# Generar datos para graficar
x_values = np.linspace(-3, 3, 100)
f_values = f_num(x_values)
derivada_values = derivada_num(x_values)

# Plotear
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label=r"$f(x)=cos(3x+3)$")   # Plotear f(x)
plt.plot(x_values, derivada_values, label=r"$f'(x)=-3sen(3x+3)$",linestyle="--")  # Plotear f'(x)

# Labels
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Escala de rango para y
plt.ylim(-5, 5)

# Título y leyendas
plt.title('Gráfica de $f(x)$ y $f\'(x)$', fontsize=16)
plt.legend(fontsize=12)

# Grilla
plt.grid(True, which='both', linestyle='--', linewidth=0.5)


plt.show()
```
