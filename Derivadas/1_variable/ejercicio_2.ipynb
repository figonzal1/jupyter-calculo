### Ejercicios derivadas 1 variable con python

2) **Calcule la derivada de la función $f(x)=2x^4+x^3-x^2+4$**

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
# Definicion de variables
x = sp.symbols('x')

# Definir la función
f = 2*x**4 + x**3 - x**2 +4

# Calcular la derivada
derivada = sp.diff(f, x)

# Resultado
display(Math(r"f'(x) = " + sp.latex(derivada)))
```

```python
# Grafica de f(x) y f'(x)

# Conversion de calculo simbolico a numerico
import numpy as np
f_num = sp.lambdify(x, f, 'numpy')
derivada_num = sp.lambdify(x, derivada, 'numpy')

# Generar datos para graficar
x_values = np.linspace(-3, 3, 100)
f_values = f_num(x_values)
derivada_values = derivada_num(x_values)

# Plotear
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label=r"$f(x)=2x^4+x^3-x^2+4$")   # Plotear f(x)
plt.plot(x_values, derivada_values, label=r"$f'(x)=8x^3+3x^2-2x$",linestyle="--")  # Plotear f'(x)

# Labels
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)

# Escala de rango para y
plt.ylim(-3, 20)

# Título y leyendas
plt.title('Gráfica de $f(x)$ y $f\'(x)$', fontsize=16)
plt.legend(fontsize=12)

# Grilla
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```
